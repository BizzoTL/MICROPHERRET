```python
import pandas as pd
```

```python
#database is a dict with classes as key. Each class contains list of organisms which belong to the class
database = {} 
file_path = '../FAPROTAX.txt' #faprotax 1.2.6
#list of the classes, len = 92
classes = []
#list of organisms, no repetitions
organisms = set()
```

```python
with open(file_path, 'r') as file:
    #assign index to the class we are referring to in classes list
    current_class_index = -1

    for line in file:
        #eliminates comments
        if line.startswith('#'):
            continue

        #separate lines
        line = line.strip().split('\t')

        #check for empty lines
        if line[0] == '':
            continue

        #save classes
        elif '*' not in line[0] and ':' not in line[0]:
            current_class_index += 1
            classes.append(line[0].replace(' ',''))
        
        #we are inside a class
        else:
            current_class = classes[current_class_index]
            key = current_class

            if not line[0].startswith('add_group:') and not line[0].startswith('subtract_group:'):
                org = line[0].replace(' ','')
                org = org.replace('*',' ').lstrip().rstrip()
                organisms.add(org)
                try: 
                    database[key].append(org)
                except KeyError:
                    database[key] = [org]
        
                
            elif line[0].startswith('add_group:'):
                key_to_add = line[0][len('add_group:'):]
                for org in database[key_to_add]:
                    try:
                        database[key].append(org)
                    except KeyError:
                        database[key] = [org]
            
            elif line[0].startswith('subtract_group:'):
                key_to_subtract = line[0][len('subtract_group:'):]
                for org in database[key_to_subtract]:
                    if org in database[key]:
                        database[key].remove(org)
        
        
```

```python
for i in database.values():
    for l in i:
        print(l)

#Results in a list of names to be processed by NCBI taxonomy name tool (https://www.ncbi.nlm.nih.gov/Taxonomy/TaxIdentifier/tax_identifier.cgi)
#That returns the ID of the organism to be used by ncbi-genome-download dry run to download the final genomes
```
